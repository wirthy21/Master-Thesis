```python
import os
import cv2
import tqdm
import pickle
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
from utils.dataset_utils import *
from collections import defaultdict
from moviepy.editor import VideoFileClip
from deep_sort_realtime.deepsort_tracker import DeepSort
```

```python
# original height: 2160

video_folder  = r"..\data\raw\pred_prey_interaction"
output_folder = r"..\data\raw\pred_prey_interaction\output"

#video_folder  = r"..\data\raw\pred_attack"
#output_folder = r"..\data\raw\pred_attack\output"

os.makedirs(output_folder, exist_ok=True)

target_w, target_h = 2160, 2160

extensions = ('.mp4', '.mov', '.avi', '.mkv')

for fname in os.listdir(video_folder):
    if not fname.lower().endswith(extensions):
        continue

    input_path  = os.path.join(video_folder, fname)
    name, _     = os.path.splitext(fname)
    output_path = os.path.join(output_folder, f"{name}.mp4")

    print(f"Verarbeite {fname} …")

    clip = VideoFileClip(input_path)
    w, h = clip.size

    x1 = (w - target_w) / 2 if w > target_w else 0
    y1 = (h - target_h) / 2 if h > target_h else 0

    cropped = clip.crop(x1=x1, y1=y1, width=target_w, height=target_h)

    cropped.write_videofile(
        output_path,
        codec="libx264",
        audio_codec="aac",
        threads=4,
        temp_audiofile="temp-audio.m4a",
        remove_temp=True,
        logger=None
    )

    clip.close()
    cropped.close()
```

```python
video_folder = r"..\data\raw\pred_attack\output"
image_folder = r"..\data\raw\pred_attack\images"

os.makedirs(image_folder, exist_ok=True)

for video in os.listdir(video_folder):
    out = os.path.join(image_folder, os.path.splitext(video)[0])
    os.makedirs(out, exist_ok=True)
    cap = cv2.VideoCapture(os.path.join(video_folder, video))
    i = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if i % 2 == 0:
            cv2.imwrite(os.path.join(out, f"frame_{i:03d}.jpg"), frame)
        i += 1
    cap.release()
```
