```python
import os
import pickle
from utils.dataset_utils import *
```

```python
total_detections = 33
num_frames = 1
processed_video_folder = r'..\data\processed\pred_prey_interactions'
prey_et_path = os.path.join(processed_video_folder, "expert_tensors", f"{total_detections}", f"prey_tensors_{num_frames}.pkl")
prey_et_save_dir = r"..\data\processed\pred_prey_interactions\expert_tensors\flagged\yolo_tensors"

with open(prey_et_path, "rb") as f:
    prey_tensors = pickle.load(f)

print("Prey Tensors Shape:", tuple(prey_tensors.shape))

B, A, N, F = prey_tensors.shape
new_tensor = torch.empty((B, A, N, F+1), dtype=prey_tensors.dtype, device=prey_tensors.device)
new_tensor[..., 0] = 0
new_tensor[..., 1:] = prey_tensors
new_tensor[:, :, 0, 0] = 1.0

prey_et_path = os.path.join(prey_et_save_dir, f"prey_tensors_flagged_{num_frames}.pkl")
with open(prey_et_path, "wb") as f:
    pickle.dump(new_tensor, f)
```
